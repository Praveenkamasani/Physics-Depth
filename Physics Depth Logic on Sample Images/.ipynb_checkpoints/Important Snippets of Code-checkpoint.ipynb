## Angle of Rotation of Rotation Matrix

```python
import numpy as np

# Function to calculate Euler angles from a rotation matrix
def rotation_matrix_to_euler_angles(R):
    assert(R.shape == (3, 3))
    
    # Calculate sy = sqrt((R[0,0] * R[0,0]) +  (R[1,0] * R[1,0]))
    sy = np.sqrt(R[0,0] ** 2 +  R[1,0] ** 2)
    singular = sy < 1e-6
    
    if not singular:  # Standard case
        x = np.arctan2(R[2,1] , R[2,2])
        y = np.arctan2(-R[2,0], sy)
        z = np.arctan2(R[1,0], R[0,0])
    else:  # Gimbal lock case
        x = np.arctan2(-R[1,2], R[1,1])
        y = np.arctan2(-R[2,0], sy)
        z = 0

    return np.degrees([x, y, z])  # Convert to degrees

# Your provided rotation matrix
R = np.array([[ 0.00775545, -0.9999694 , -0.0010143 ],
              [ 0.00229406,  0.00103212, -0.9999968 ],
              [ 0.9999673 ,  0.0077531 ,  0.00230199]])

# Extract the Euler angles
euler_angles = rotation_matrix_to_euler_angles(R)

# Print the Euler angles
print(f"Rotation about the x-axis (gamma): {euler_angles[0]:.2f} degrees")
print(f"Rotation about the y-axis (beta): {euler_angles[1]:.2f} degrees")
print(f"Rotation about the z-axis (alpha): {euler_angles[2]:.2f} degrees")
```

## Visualization of Angle of Rotation of Rotation Matrix

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Your provided rotation matrix
R = np.array([[ 0.00775545, -0.9999694 , -0.0010143 ],
              [ 0.00229406,  0.00103212, -0.9999968 ],
              [ 0.9999673 ,  0.0077531 ,  0.00230199]])

# Set up the figure and 3D axis
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Original axes
axes_info = [
    {'start': [0, 0, 0], 'vector': [1, 0, 0], 'color': 'r', 'label': 'X0'},
    {'start': [0, 0, 0], 'vector': [0, 1, 0], 'color': 'g', 'label': 'Y0'},
    {'start': [0, 0, 0], 'vector': [0, 0, 1], 'color': 'b', 'label': 'Z0'}
]

# Function to draw axes
def draw_axes(ax, axes_info, R=None):
    for info in axes_info:
        # Apply rotation if provided
        vector = info['vector'] if R is None else R @ info['vector']
        ax.quiver(*info['start'], *vector, length=1.0, color=info['color'])
        ax.text(vector[0], vector[1], vector[2], info['label'], color=info['color'], size=20)

# Plot original axes
draw_axes(ax, axes_info)

# Plot rotated axes
rotated_axes_info = [{'start': [0, 0, 0], 'vector': v_rotated, 'color': c, 'label': l + "'"} 
                     for v_rotated, c, l in [(R @ info['vector'], info['color'], info['label']) for info in axes_info]]
draw_axes(ax, rotated_axes_info)

# Set the aspect ratio to 'auto' to avoid skewing the plot
ax.set_aspect('auto')

# Set labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Rotation Visualization')

# Set axis limits
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.5, 1.5])

# Show the grid and the plot
ax.grid(True)

print()
plt.show()
```

## Metrics Calculation Code

```python
import numpy as np

def calculate_depth_metrics(predicted_depth_map, ground_truth_depth_map, depth_threshold=80):
    # Ensure calculations are only done where both depths are non-zero and predicted depth is less than the threshold
    mask = (ground_truth_depth_map > 0) & (predicted_depth_map > 0) & (predicted_depth_map < depth_threshold) 
    valid_pixel_count = np.sum(mask)
    predicted_depth_map = predicted_depth_map[mask]
    ground_truth_depth_map = ground_truth_depth_map[mask]

    # Avoid division by zero and log of zero
    if np.any(ground_truth_depth_map <= 0) or np.any(predicted_depth_map <= 0):
        raise ValueError("Depth values should be positive and non-zero.")

    # Calculate errors
    error = np.abs(ground_truth_depth_map - predicted_depth_map)
    percentage_error = np.mean(error / ground_truth_depth_map) * 100

    # Metrics
    abs_rel = np.mean(np.abs(ground_truth_depth_map - predicted_depth_map) / ground_truth_depth_map)
    sq_rel = np.mean(((ground_truth_depth_map - predicted_depth_map) ** 2) / ground_truth_depth_map)
    rmse = np.sqrt(np.mean((ground_truth_depth_map - predicted_depth_map) ** 2))
    rmse_log = np.sqrt(np.mean((np.log(ground_truth_depth_map) - np.log(predicted_depth_map)) ** 2))
    delta1 = (np.maximum((ground_truth_depth_map / predicted_depth_map), (predicted_depth_map / ground_truth_depth_map)) < 1.25).mean()
    delta2 = (np.maximum((ground_truth_depth_map / predicted_depth_map), (predicted_depth_map / ground_truth_depth_map)) < 1.25 ** 2).mean()
    delta3 = (np.maximum((ground_truth_depth_map / predicted_depth_map), (predicted_depth_map / ground_truth_depth_map)) < 1.25 ** 3).mean()

    # Additional metrics
    max_error = np.max(error)
    min_error = np.min(error)
    mean_error = np.mean(error)
    abs_mean_error = np.mean(np.abs(error))
    predicted_depth_range = (np.min(predicted_depth_map), np.max(predicted_depth_map))
    ground_truth_depth_range = (np.min(ground_truth_depth_map), np.max(ground_truth_depth_map))

    return {
        "Valid Pixel Count": valid_pixel_count,
        "AbsRel": abs_rel,
        "Sq Rel": sq_rel,
        "RMSE": rmse,
        "RMSE log": rmse_log,
        "Î´ < 1.25": delta1,
        "Î´ < 1.25^2": delta2,
        "Î´ < 1.25^3": delta3,
        "Max Error": max_error,
        "Min Error": min_error,
        "Mean Error": mean_error,
        "Percentage Error": percentage_error,
        "Predicted Depth Range": predicted_depth_range,
        "Ground Truth Depth Range": ground_truth_depth_range
    }

#predicted_depth_map = 
#ground_truth_depth_map = 

metrics = calculate_depth_metrics(predicted_depth_map, ground_truth_depth_map)
# Print all metrics
for metric, value in metrics.items():
    print(f"{metric}: {value}")
```
