```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load Disparity Map
img_disp_name = '/home/pk11519/Downloads/Cityscape Sample File/aachen_000002_000019_disparity.png'  # Modify with your image path
img_d = cv2.imread(img_disp_name, cv2.IMREAD_UNCHANGED).astype(np.float32)
# Check for pixels with disparity values greater than zero and normalize
img_d[img_d > 0] = (img_d[img_d > 0] - 1) / 256
# Calculate depth map where disparity is not zero, else depth is zero
focal_length = 2262.52  # example focal length in pixels
baseline = 0.209313     # example baseline in meters
depth = np.zeros(img_d.shape, dtype=np.float32)  #Initialize depth map with zeros
non_zero_disparity = img_d > 0
depth[non_zero_disparity] = (focal_length * baseline) / img_d[non_zero_disparity]

# Visualize depth map
fig = plt.figure(dpi=350)
plt.axis('off')
plt.imshow(depth)
plt.colorbar(shrink=0.35)
plt.show()

masked_image = np.ma.masked_where((depth == 0)|(depth==0), depth) # For Visualization Purpose
fig = plt.figure(dpi=350)
plt.axis('off')
plt.imshow(masked_image)
plt.colorbar(shrink=0.35)
plt.show()

# Calculate min, max, mean, median of non-zero depths
valid_depths = depth[non_zero_disparity]  # Filter out zero depth values
min_depth = np.min(valid_depths)
max_depth = np.max(valid_depths)
mean_depth = np.mean(valid_depths)
median_depth = np.median(valid_depths)

# Print depth statistics for non-zero values
print(f'Min depth (for non-zero disparity): {min_depth} mtrs')
print(f'Mean depth (for non-zero disparity): {mean_depth} mtrs')
print(f'Median depth (for non-zero disparity): {median_depth} mtrs')
print(f'Max depth (for non-zero disparity): {max_depth} mtrs')
```

```python
import numpy as np

def create_rotation_matrix(r, p, y):
    # Rotation matrix around X-axis (Roll)
    R_x = np.array([[1, 0, 0],
                    [0, np.cos(r), -np.sin(r)],
                    [0, np.sin(r), np.cos(r)]])
    # Rotation matrix around Y-axis (Pitch)
    R_y = np.array([[np.cos(p), 0, np.sin(p)],
                    [0, 1, 0],
                    [-np.sin(p), 0, np.cos(p)]])
    # Rotation matrix around Z-axis (Yaw)
    R_z = np.array([[np.cos(y), -np.sin(y), 0],
                    [np.sin(y), np.cos(y), 0],
                    [0, 0, 1]])
    # Combined rotation matrix
    R = np.dot(R_z, np.dot(R_y, R_x))
    return R

# Example usage
roll =  0.0  # Roll angle in degrees
pitch = 0.038  # Pitch angle in degrees
yaw = -0.0195  # Yaw angle in degrees

R_CtoV = create_rotation_matrix(roll, pitch, yaw)
R_VtoC=np.linalg.inv(R_CtoV)
```

```python
K = np.array([
    [ 2262.52, 0.00,                1096.98], 
    [0,        2265.3017905988554 , 513.137], 
    [0.00,     0.00,                1.00]
])
print("K", type(K))
print("K", K)

R = np.array([
    [0, -1,  0], 
    [0,  0, -1], 
    [1,  0,  0]
])
print("R", type(R))
print("R", R)

#A_proj=np.dot(K,R)

A_proj=np.dot(np.dot(K,R), R_VtoC)
print("A_proj", type(A_proj))
print("A_proj", A_proj)
```

```python
# project to image
phy_depth = np.zeros((1024, 2048))

# Calculate depth for each pixel
for y in range(1024):
    for x in range(2048):
        u=x
        v=y
        # Original equations
        equations = [
            {"coefficients": [u, -A_proj[0,0], -A_proj[0,1]], "rhs":A_proj[0,2]*(-1.22)},  # u*z_c - 5*x_w - 6*y_w = 4
            {"coefficients": [v, -A_proj[1,0], -A_proj[1,1]], "rhs":A_proj[1,2]*(-1.22)},  # v*z_c - 7*x_w - 8*y_w = 6
            {"coefficients": [1, -A_proj[2,0], -A_proj[2,1]], "rhs":A_proj[2,2]*(-1.22)}   # z_c   - 5*x_w - 4*y_w = 3
        ]

        # Forming the coefficient matrix and right-hand side vector
        A = np.array([eq["coefficients"] for eq in equations])
        B = np.array([eq["rhs"] for eq in equations])

        # Solving the system of equations using numpy's linear algebra solver
        solution = np.linalg.solve(A, B)

        # Extracting the values of z_c, x_w, and y_w from the solution
        z_c, x_w, y_w = solution
        
        phy_depth[y,x] = x_w
        #phy_depth[y,x] = np.sqrt(x_w**2 + y_w**2 + (-1.22)**2)

phy_depth_map_full_image= phy_depth
```

```python
import os
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

# Initialize frequency counter
freq_counter = Counter()

# Load Physics Depth
physics_depth_map = phy_depth_map_road
ld_depth_map = depth_road

# Create an empty array with the same shape as the original depth maps
percentage_difference_map = np.zeros_like(ld_depth_map)

# Iterate over the depth maps
for i in range(ld_depth_map.shape[0]):
    for j in range(ld_depth_map.shape[1]):
        lidar_value = ld_depth_map[i, j]
        physics_value = physics_depth_map[i, j]

        # Check if either value is zero
        if lidar_value == 0 or physics_value == 0:
            percentage_difference = 0
        else:
            # Calculate the percentage difference and round to 2 decimals
            percentage_difference = round(((lidar_value-physics_value) / (lidar_value)) * 100, 2)
            
        # Place the calculated value in the percentage difference map
        percentage_difference_map[i, j] = percentage_difference

# Flatten the array to work with 1D data
flattened_map = percentage_difference_map.flatten()
# Filter out zero values
non_zero_values = flattened_map[flattened_map != 0]

# Update frequency counter
for val in non_zero_values:
    if -50 < val < 50:
        freq_counter[int(val)] += 1

# Print and save frequency
total_count = sum(freq_counter.values())

for i in range(-50, 50):
    count = freq_counter.get(i, 0)
    #print(f"Range ({i}, {i+1}): {count} values")

# Plot histogram
plt.figure(figsize=(15, 8))  # Increase the size of the histogram
plt.hist(list(freq_counter.elements()), bins=range(-50, 51), edgecolor='black', density=True)

# Convert y-axis to percentage
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(xmax=1))

plt.xlabel('Value')
plt.ylabel('Percentage (%)')
plt.title('Frequency Histogram')
plt.show()
```

### Metrics Calculations 

```python
import numpy as np

def calculate_depth_metrics(predicted_depth_map, ground_truth_depth_map, depth_threshold=80):
    # Ensure calculations are only done where both depths are non-zero and predicted depth is less than the threshold
    mask = (ground_truth_depth_map > 0) & (predicted_depth_map > 0) & (predicted_depth_map < depth_threshold) & (ground_truth_depth_map < depth_threshold)
    valid_pixel_count = np.sum(mask)
    predicted_depth_map = predicted_depth_map[mask]
    ground_truth_depth_map = ground_truth_depth_map[mask]

    # Avoid division by zero and log of zero
    if np.any(ground_truth_depth_map <= 0) or np.any(predicted_depth_map <= 0):
        raise ValueError("Depth values should be positive and non-zero.")

    # Calculate errors
    error = np.abs(ground_truth_depth_map - predicted_depth_map)
    percentage_error = np.mean(error / ground_truth_depth_map) * 100

    # Metrics
    abs_rel = np.mean(np.abs(ground_truth_depth_map - predicted_depth_map) / ground_truth_depth_map)
    sq_rel = np.mean(((ground_truth_depth_map - predicted_depth_map) ** 2) / ground_truth_depth_map)
    rmse = np.sqrt(np.mean((ground_truth_depth_map - predicted_depth_map) ** 2))
    rmse_log = np.sqrt(np.mean((np.log(ground_truth_depth_map) - np.log(predicted_depth_map)) ** 2))
    delta1 = (np.maximum((ground_truth_depth_map / predicted_depth_map), (predicted_depth_map / ground_truth_depth_map)) < 1.25).mean()
    delta2 = (np.maximum((ground_truth_depth_map / predicted_depth_map), (predicted_depth_map / ground_truth_depth_map)) < 1.25 ** 2).mean()
    delta3 = (np.maximum((ground_truth_depth_map / predicted_depth_map), (predicted_depth_map / ground_truth_depth_map)) < 1.25 ** 3).mean()

    # Additional metrics
    max_error = np.max(error)
    min_error = np.min(error)
    mean_error = np.mean(error)
    abs_mean_error = np.mean(np.abs(error))
    predicted_depth_range = (np.min(predicted_depth_map), np.max(predicted_depth_map))
    ground_truth_depth_range = (np.min(ground_truth_depth_map), np.max(ground_truth_depth_map))

    return {
        "Valid Pixel Count": valid_pixel_count,
        "AbsRel": abs_rel,
        "Sq Rel": sq_rel,
        "RMSE": rmse,
        "RMSE log": rmse_log,
        " < 1.25": delta1,
        " < 1.25^2": delta2,
        " < 1.25^3": delta3,
        "Max Error": max_error,
        "Min Error": min_error,
        "Mean Error": mean_error,
        "Percentage Error": percentage_error,
        "Predicted Depth Range": predicted_depth_range,
        "Ground Truth Depth Range": ground_truth_depth_range
    }

predicted_depth_map = phy_depth_map_road
ground_truth_depth_map = depth_road

metrics = calculate_depth_metrics(predicted_depth_map, ground_truth_depth_map)

# Print all metrics
for metric, value in metrics.items():
    print(f"{metric}: {value}")
```
